## Section 2: NumPy Essentials

NumPy gives us arrays — fast, vectorized containers for numerical data. Think of it as an Excel column on steroids.

**Biomedical context**: Imagine you have 10,000 patient blood pressure readings. NumPy lets you compute the mean, standard deviation, and percentiles in one line — no loops needed.

````python
# Creating arrays
lab_values = np.array([5.2, 4.8, 6.1, 5.5, 4.9, 5.0, 7.3, 4.6])
print(f'Array: {lab_values}')
print(f'Type: {lab_values.dtype}')
print(f'Length: {len(lab_values)}')
````

````python
# Vectorized operations — no loops!
# Convert mmol/L to mg/dL (multiply by 18.018 for glucose)
lab_values_mgdl = lab_values * 18.018
print(f'mmol/L: {lab_values}')
print(f'mg/dL:  {np.round(lab_values_mgdl, 1)}')
````

````python
# Statistical functions
print(f'Mean:   {np.mean(lab_values):.2f} mmol/L')
print(f'Median: {np.median(lab_values):.2f} mmol/L')
print(f'Std:    {np.std(lab_values):.2f} mmol/L')
print(f'Min:    {np.min(lab_values):.2f} mmol/L')
print(f'Max:    {np.max(lab_values):.2f} mmol/L')
print(f'25th percentile: {np.percentile(lab_values, 25):.2f} mmol/L')
print(f'75th percentile: {np.percentile(lab_values, 75):.2f} mmol/L')
````

````python
# Boolean filtering
normal_mask = lab_values <= 5.5       # Normal fasting glucose <= 5.5 mmol/L
high_mask = lab_values > 5.5
print(f'Normal readings: {lab_values[normal_mask]}')
print(f'High readings:   {lab_values[high_mask]}')
print(f'Count high: {np.sum(high_mask)} out of {len(lab_values)}')
````

## Section 3: Pandas Fundamentals

Pandas is the workhorse of data analysis in Python. It gives us the **DataFrame** — think of it as a programmable spreadsheet.

### Our Dataset: Daily Live Births in Malaysia
- Source: National Registration Department (JPN) via [data.gov.my](https://data.gov.my)
- Columns: `date`, `state`, `births`
- Time range: 1920 to 2023

````python
# Load the dataset directly from the web
BIRTHS_URL = 'https://storage.data.gov.my/demography/births.csv'
df = pd.read_csv(BIRTHS_URL)
print(f'Loaded {len(df):,} rows')
````

````python
# First look at the data
df.head(10)
````

````python
# Structure overview
print('--- Column info ---')
df.info()
print('\n--- Summary statistics ---')
df.describe()
````

````python
# Check for missing values
print('Missing values per column:')
df.isnull().sum()
````

### Working with Dates
The `date` column is currently read as text. Let's convert it to proper datetime objects so we can extract year, month, day of week, etc.

````python
# Convert to datetime
df['date'] = pd.to_datetime(df['date'])

# Extract date components
df['year'] = df['date'].dt.year
df['month'] = df['date'].dt.month
df['day'] = df['date'].dt.day
df['dayofweek'] = df['date'].dt.dayofweek  # Monday=0, Sunday=6
df['day_name'] = df['date'].dt.day_name()
df['month_name'] = df['date'].dt.month_name()

df.head()
````

````python
# Filtering: births after the year 2000
df_recent = df[df['year'] >= 2000]
print(f'Rows since 2000: {len(df_recent):,}')

# Filtering: days with more than 2,000 births
df_high = df[df['births'] > 2000]
print(f'Days with >2,000 births: {len(df_high)}')
df_high.head()
````

````python
# Grouping: total births per year
yearly = df.groupby('year')['births'].sum().reset_index()
yearly.head(10)
````

````python
# Multiple aggregations at once
yearly_stats = df.groupby('year').agg(
    total_births=('births', 'sum'),
    avg_daily=('births', 'mean'),
    max_daily=('births', 'max'),
    days_recorded=('births', 'count')
).reset_index()
yearly_stats.head()
````

## Section 4: Data Visualization

We'll use **Matplotlib** (low-level, full control) and **Seaborn** (high-level, statistical plots).
Both are pre-installed in Colab.

The golden rule: **always label your axes and add a title** — your plots should tell a story without you explaining them.

````python
# Line plot: Total births per year
fig, ax = plt.subplots(figsize=(12, 5))
ax.plot(yearly['year'], yearly['births'], linewidth=1.5, color='steelblue')
ax.set_title('Total Live Births in Malaysia per Year (1920-2023)', fontsize=14, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Total Births')
ax.set_ylim(bottom=0)
plt.tight_layout()
plt.show()
````

🤔 **What do you notice?** There's a dramatic drop around 1942-1945 (WWII Japanese occupation of Malaya) and a post-war baby boom.

````python
# Bar plot: Average daily births by month
monthly_avg = df.groupby('month')['births'].mean().reset_index()
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.bar(month_names, monthly_avg['births'], color='steelblue', edgecolor='white')
ax.set_title('Average Daily Births by Month', fontsize=14, fontweight='bold')
ax.set_xlabel('Month')
ax.set_ylabel('Avg Daily Births')
# Highlight the bar with highest value
max_idx = monthly_avg['births'].idxmax()
bars[max_idx].set_color('coral')
plt.tight_layout()
plt.show()
````

````python
# Histogram: Distribution of daily births
fig, ax = plt.subplots(figsize=(10, 5))
ax.hist(df['births'], bins=50, color='steelblue', edgecolor='white', alpha=0.8)
ax.set_title('Distribution of Daily Births', fontsize=14, fontweight='bold')
ax.set_xlabel('Number of Births per Day')
ax.set_ylabel('Frequency (number of days)')
ax.axvline(df['births'].median(), color='coral', linestyle='--', linewidth=2,
           label=f"Median: {df['births'].median():.0f}")
ax.legend()
plt.tight_layout()
plt.show()
````

````python
# Box plot: Births by day of week (seaborn)
day_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

fig, ax = plt.subplots(figsize=(10, 5))
sns.boxplot(data=df, x='day_name', y='births', order=day_order,
            palette='Blues', ax=ax)
ax.set_title('Daily Births by Day of Week', fontsize=14, fontweight='bold')
ax.set_xlabel('Day of Week')
ax.set_ylabel('Number of Births')
plt.tight_layout()
plt.show()
````

🤔 **Notice the weekend drop!** This is a well-known phenomenon — fewer births on Saturdays and Sundays, largely due to scheduled C-sections and inductions being planned for weekdays. A clear biomedical signal in the data.

**Why it matters**: Do fewer babies arrive on public holidays? This would strengthen the argument for medical scheduling effects — since elective procedures are rarely scheduled on holidays.
